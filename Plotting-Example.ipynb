# ðŸ“Š Data Visualization: Simple â†’ Advanced (Step-by-Step Learning)
## ðŸ“¦ 1. Setup & First Plot (Super Simple)

```python
# Step 1: Import libraries
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Create simple data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Step 3: Create your first plot
plt.plot(x, y)  # Line plot
plt.title("My First Plot")  # Add title
plt.xlabel("X Axis")  # Add x-axis label
plt.ylabel("Y Axis")  # Add y-axis label
plt.show()  # Show the plot

print("âœ… Created first plot!")
```

# ðŸ“ˆ 2. Basic Plot Types (Foundation)
## 2.1 Line Plot

```python
# Line plot - For trends over time
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [100, 120, 90, 150, 200, 180]

plt.figure(figsize=(10, 6))  # Set figure size
plt.plot(months, sales, marker='o', color='blue', linewidth=2, markersize=8)
plt.title('Monthly Sales Trend', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales ($)', fontsize=12)
plt.grid(True, alpha=0.3)  # Add grid
plt.xticks(rotation=45)  # Rotate x labels
plt.show()
```

# 2.2 Bar Chart

```python
# Bar chart - For comparing categories
products = ['Laptop', 'Phone', 'Tablet', 'Headphones']
units_sold = [1500000, 300000000, 1000000, 25021200]

plt.figure(figsize=(10, 6))
bars = plt.bar(products, units_sold, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])

# Customize bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 5, 
             f'{height}', ha='center', va='bottom')

plt.title('Product Sales', fontsize=16)
plt.xlabel('Products', fontsize=12)
plt.ylabel('Units Sold', fontsize=12)
plt.ylim(0, max(units_sold) + 50)  # Set y-axis limit
plt.show()
```

## 2.3 Scatter Plot

```python
# Scatter plot - For relationships between variables
np.random.seed(42)
hours_studied = np.random.randint(1, 10, 50)
exam_scores = hours_studied * 10 + np.random.normal(0, 5, 50)

plt.figure(figsize=(10, 6))
plt.scatter(hours_studied, exam_scores, alpha=0.6, c=exam_scores, cmap='viridis', s=100)

# Add trend line
z = np.polyfit(hours_studied, exam_scores, 1)
p = np.poly1d(z)
plt.plot(hours_studied, p(hours_studied), "r--", alpha=0.8, linewidth=2)

plt.title('Study Hours vs Exam Scores', fontsize=16)
plt.xlabel('Hours Studied', fontsize=12)
plt.ylabel('Exam Score', fontsize=12)
plt.colorbar(label='Score')  # Add color bar
plt.show()
```

# 2.4 Histogram

```python
# Histogram - For distribution of data
np.random.seed(42)
ages = np.random.normal(35, 10, 1000)  # 1000 people, mean age 35, std 10

plt.figure(figsize=(10, 6))
plt.hist(ages, bins=30, edgecolor='black', alpha=0.7, color='lightblue')

# Add mean line
plt.axvline(ages.mean(), color='red', linestyle='dashed', linewidth=2, 
            label=f'Mean: {ages.mean():.1f} years')

plt.title('Age Distribution of Customers', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.legend()
plt.show()
```

# ðŸ”„ 3. Multiple Plots in One Figure

```python
# Create 2x2 grid of different plot types
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Data Analysis Dashboard', fontsize=16, fontweight='bold')

# Data for all plots
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue = [100, 150, 120, 200, 180, 220]
customers = [50, 70, 60, 90, 85, 95]
products = ['A', 'B', 'C', 'D']
sales = [120, 85, 150, 95]
age_data = np.random.normal(30, 8, 200)

# Plot 1: Line chart
axes[0, 0].plot(months, revenue, marker='o', color='blue')
axes[0, 0].set_title('Monthly Revenue')
axes[0, 0].set_ylabel('Revenue ($)')
axes[0, 0].grid(True, alpha=0.3)

# Plot 2: Bar chart
axes[0, 1].bar(products, sales, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
axes[0, 1].set_title('Product Sales')
axes[0, 1].set_ylabel('Units Sold')

# Plot 3: Scatter plot
axes[1, 0].scatter(customers, revenue, alpha=0.6, color='green')
axes[1, 0].set_title('Customers vs Revenue')
axes[1, 0].set_xlabel('Customers')
axes[1, 0].set_ylabel('Revenue ($)')

# Plot 4: Histogram
axes[1, 1].hist(age_data, bins=20, edgecolor='black', alpha=0.7, color='orange')
axes[1, 1].set_title('Customer Age Distribution')
axes[1, 1].set_xlabel('Age')
axes[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## ðŸŽ¨ 4. Introduction to Seaborn (Easier & Prettier)

```python
import seaborn as sns

# Load built-in dataset
tips = sns.load_dataset('tips')
print("Sample data from 'tips' dataset:")
print(tips.head())

# Simple Seaborn plots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. Count plot (categorical counts)
sns.countplot(data=tips, x='day', ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Visits by Day')

# 2. Box plot (distribution by category)
sns.boxplot(data=tips, x='day', y='total_bill', ax=axes[0, 1], palette='Set2')
axes[0, 1].set_title('Total Bill by Day')

# 3. Histogram with KDE
sns.histplot(data=tips, x='total_bill', kde=True, ax=axes[1, 0], color='skyblue')
axes[1, 0].set_title('Distribution of Total Bill')

# 4. Scatter plot with hue
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', 
                style='smoker', ax=axes[1, 1], s=100)
axes[1, 1].set_title('Tip vs Total Bill')

plt.tight_layout()
plt.show()
```

# ðŸ“Š 5. Business Dashboard Example

```python
# ðŸ“Š COMPLETE BUSINESS DASHBOARD EXAMPLE (Fixed)

# Step 1: Import ALL required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Set visualization style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['font.size'] = 12

# Step 3: Create a sales dashboard
np.random.seed(42)
dates = pd.date_range('2024-01-01', periods=90, freq='D')

# Create sales data
sales_data = pd.DataFrame({
    'date': dates,
    'product_a': np.random.normal(100, 20, 90).cumsum(),
    'product_b': np.random.normal(80, 15, 90).cumsum(),
    'product_c': np.random.normal(120, 25, 90).cumsum(),
    'visitors': np.random.randint(500, 1500, 90),
    'conversion_rate': np.random.uniform(0.01, 0.05, 90)
})

# Calculate weekly data
sales_data['week'] = sales_data['date'].dt.isocalendar().week
weekly_sales = sales_data.groupby('week').agg({
    'product_a': 'sum',
    'product_b': 'sum', 
    'product_c': 'sum',
    'visitors': 'mean'
}).reset_index()

print("ðŸ“Š Sales Data Preview:")
print(sales_data.head())
print(f"\nðŸ“ˆ Dataset Shape: {sales_data.shape}")
print(f"ðŸ“‹ Columns: {', '.join(sales_data.columns.tolist())}")

# Step 4: Create dashboard
fig = plt.figure(figsize=(16, 12))
fig.suptitle('ðŸ“ˆ Sales Performance Dashboard\nQ1 2024', fontsize=18, fontweight='bold')

# Grid layout
gs = fig.add_gridspec(3, 3, hspace=0.4, wspace=0.3)

# 1. Line plot - Daily sales
ax1 = fig.add_subplot(gs[0, :2])
ax1.plot(sales_data['date'], sales_data['product_a'], label='Product A', linewidth=2)
ax1.plot(sales_data['date'], sales_data['product_b'], label='Product B', linewidth=2)
ax1.plot(sales_data['date'], sales_data['product_c'], label='Product C', linewidth=2)
ax1.set_title('Daily Sales Trend')
ax1.set_xlabel('Date')
ax1.set_ylabel('Sales ($)')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.tick_params(axis='x', rotation=45)

# 2. Bar plot - Weekly comparison
ax2 = fig.add_subplot(gs[0, 2])
weeks = weekly_sales['week'][:4]  # First 4 weeks
product_a_sales = weekly_sales['product_a'][:4]
product_b_sales = weekly_sales['product_b'][:4]
product_c_sales = weekly_sales['product_c'][:4]

x = np.arange(len(weeks))
width = 0.25

bars1 = ax2.bar(x - width, product_a_sales, width, label='Product A', color='skyblue')
bars2 = ax2.bar(x, product_b_sales, width, label='Product B', color='lightcoral')
bars3 = ax2.bar(x + width, product_c_sales, width, label='Product C', color='lightgreen')

ax2.set_title('Weekly Sales Comparison')
ax2.set_xlabel('Week')
ax2.set_ylabel('Sales ($)')
ax2.set_xticks(x)
ax2.set_xticklabels([f'Week {w}' for w in weeks])
ax2.legend()

# Add value labels on bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2, height, 
                f'${height:,.0f}', ha='center', va='bottom', fontsize=8)

add_labels(bars1)
add_labels(bars2)
add_labels(bars3)

# 3. Pie chart - Product mix
ax3 = fig.add_subplot(gs[1, 0])
total_sales = [sales_data['product_a'].sum(), 
               sales_data['product_b'].sum(), 
               sales_data['product_c'].sum()]
colors = ['#ff9999', '#66b3ff', '#99ff99']
wedges, texts, autotexts = ax3.pie(total_sales, labels=['A', 'B', 'C'], 
                                   autopct='%1.1f%%', colors=colors, startangle=90)
ax3.set_title('Product Sales Mix')
ax3.axis('equal')  # Equal aspect ratio ensures pie is drawn as circle

# 4. Scatter plot - Visitors vs Sales
ax4 = fig.add_subplot(gs[1, 1])
scatter = ax4.scatter(sales_data['visitors'], sales_data['product_a'], 
                      c=sales_data['conversion_rate']*100, cmap='RdYlGn',
                      alpha=0.6, s=50)
ax4.set_title('Visitors vs Sales (Product A)')
ax4.set_xlabel('Daily Visitors')
ax4.set_ylabel('Sales ($)')
plt.colorbar(scatter, ax=ax4, label='Conversion Rate (%)')

# 5. Heatmap - Correlation
ax5 = fig.add_subplot(gs[1, 2])
correlation = sales_data[['product_a', 'product_b', 'product_c', 'visitors']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, ax=ax5, fmt='.2f')
ax5.set_title('Feature Correlation Heatmap')

# 6. Area chart - Cumulative sales
ax6 = fig.add_subplot(gs[2, :])
ax6.fill_between(sales_data['date'], 0, sales_data['product_a'], alpha=0.3, label='Product A', color='skyblue')
ax6.fill_between(sales_data['date'], sales_data['product_a'], 
                 sales_data['product_a'] + sales_data['product_b'], 
                 alpha=0.3, label='Product B', color='lightcoral')
ax6.fill_between(sales_data['date'], 
                 sales_data['product_a'] + sales_data['product_b'],
                 sales_data['product_a'] + sales_data['product_b'] + sales_data['product_c'],
                 alpha=0.3, label='Product C', color='lightgreen')
ax6.set_title('Cumulative Sales Over Time')
ax6.set_xlabel('Date')
ax6.set_ylabel('Cumulative Sales ($)')
ax6.legend(loc='upper left')
ax6.grid(True, alpha=0.3)
ax6.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print("âœ… Dashboard created successfully!")
```

# ðŸŽ¯ 6. Advanced Visualization Techniques
## 6.1 Interactive Hover Effects

```python
# Create interactive-like plot with annotations
np.random.seed(42)
x = np.arange(1, 31)
y1 = np.random.rand(30) * 100
y2 = np.random.rand(30) * 100 + 50

fig, ax = plt.subplots(figsize=(14, 7))

# Plot lines
line1, = ax.plot(x, y1, 'o-', color='blue', linewidth=2, markersize=8, label='Team A')
line2, = ax.plot(x, y2, 's-', color='red', linewidth=2, markersize=8, label='Team B')

# Highlight max points
max_y1_idx = np.argmax(y1)
max_y2_idx = np.argmax(y2)

ax.annotate(f'Max: {y1[max_y1_idx]:.0f}', 
            xy=(x[max_y1_idx], y1[max_y1_idx]),
            xytext=(x[max_y1_idx]+1, y1[max_y1_idx]+10),
            arrowprops=dict(arrowstyle='->', color='blue'),
            fontsize=10, color='blue')

ax.annotate(f'Max: {y2[max_y2_idx]:.0f}', 
            xy=(x[max_y2_idx], y2[max_y2_idx]),
            xytext=(x[max_y2_idx]+1, y2[max_y2_idx]+10),
            arrowprops=dict(arrowstyle='->', color='red'),
            fontsize=10, color='red')

# Add shaded area between lines
ax.fill_between(x, y1, y2, where=(y2 > y1), color='green', alpha=0.2, label='Team B Lead')
ax.fill_between(x, y1, y2, where=(y1 > y2), color='orange', alpha=0.2, label='Team A Lead')

ax.set_title('Team Performance Comparison', fontsize=16, fontweight='bold')
ax.set_xlabel('Day', fontsize=12)
ax.set_ylabel('Score', fontsize=12)
ax.legend()
ax.grid(True, alpha=0.3)

plt.show()
```

## 6.2 Subplot with Different Scales

```python
# Dual y-axis plot
fig, ax1 = plt.subplots(figsize=(12, 6))

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue = [100, 150, 120, 200, 180, 220]
customers = [50, 70, 60, 90, 85, 95]

color1 = 'tab:blue'
ax1.set_xlabel('Month')
ax1.set_ylabel('Revenue ($)', color=color1)
line1 = ax1.plot(months, revenue, color=color1, marker='o', linewidth=2, label='Revenue')
ax1.tick_params(axis='y', labelcolor=color1)

# Create second y-axis
ax2 = ax1.twinx()
color2 = 'tab:red'
ax2.set_ylabel('Customers', color=color2)
line2 = ax2.plot(months, customers, color=color2, marker='s', linewidth=2, label='Customers')
ax2.tick_params(axis='y', labelcolor=color2)

# Add title and combine legends
plt.title('Revenue vs Customers', fontsize=16, fontweight='bold')

# Combine legends
lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper left')

plt.grid(True, alpha=0.3)
plt.show()
```

## 6.3 Violin Plot for Distribution

```python
# Create sample data
np.random.seed(42)
data_dict = {
    'Group A': np.random.normal(50, 10, 100),
    'Group B': np.random.normal(60, 15, 100),
    'Group C': np.random.normal(40, 5, 100),
    'Group D': np.random.normal(70, 20, 100)
}

# Convert to DataFrame
df = pd.DataFrame(data_dict).melt(var_name='Group', value_name='Value')

# Create violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Group', y='Value', palette='Set2', inner='quartile')

# Add mean points
means = df.groupby('Group')['Value'].mean()
for i, (group, mean_val) in enumerate(means.items()):
    plt.scatter(i, mean_val, color='red', s=100, zorder=3, label='Mean' if i == 0 else "")

plt.title('Distribution Comparison with Violin Plot', fontsize=16, fontweight='bold')
plt.xlabel('Group')
plt.ylabel('Value')
plt.legend()
plt.grid(True, alpha=0.3, axis='y')
plt.show()
```

## ðŸ“‹ 7. Complete Reporting Template

```python
def create_business_report(data):
    """Create a complete business report with visualizations"""
    
    fig = plt.figure(figsize=(18, 14))
    fig.suptitle('ðŸ“Š Monthly Business Performance Report\nQ1 2024', 
                 fontsize=20, fontweight='bold', y=1.02)
    
    # Define grid
    gs = fig.add_gridspec(3, 3, hspace=0.4, wspace=0.3)
    
    # 1. KPI Summary
    ax1 = fig.add_subplot(gs[0, 0])
    kpis = {
        'Total Revenue': f'${data["revenue"].sum():,.0f}',
        'Avg. Order': f'${data["revenue"].mean():,.0f}',
        'Growth Rate': f'{((data["revenue"].iloc[-1] / data["revenue"].iloc[0]) - 1)*100:+.1f}%',
        'Customers': f'{data["customers"].sum():,.0f}'
    }
    
    ax1.axis('off')
    ax1.text(0.1, 0.9, 'Key Performance Indicators', 
             fontsize=14, fontweight='bold', transform=ax1.transAxes)
    
    for i, (kpi, value) in enumerate(kpis.items()):
        y_pos = 0.7 - i * 0.15
        ax1.text(0.1, y_pos, f'{kpi}:', fontsize=12, fontweight='bold', 
                 transform=ax1.transAxes)
        ax1.text(0.6, y_pos, value, fontsize=12, color='blue',
                 fontweight='bold', transform=ax1.transAxes)
    
    # 2. Revenue Trend
    ax2 = fig.add_subplot(gs[0, 1:])
    ax2.plot(data['month'], data['revenue'], 'o-', linewidth=2, markersize=8)
    ax2.fill_between(data['month'], 0, data['revenue'], alpha=0.3)
    ax2.set_title('Revenue Trend', fontsize=14, fontweight='bold')
    ax2.set_ylabel('Revenue ($)')
    ax2.grid(True, alpha=0.3)
    ax2.tick_params(axis='x', rotation=45)
    
    # 3. Customer Metrics
    ax3 = fig.add_subplot(gs[1, 0])
    metrics = ['New', 'Returning', 'Churned']
    values = [data['new_customers'].sum(), 
              data['returning_customers'].sum(), 
              data['churned_customers'].sum()]
    
    bars = ax3.bar(metrics, values, color=['green', 'blue', 'red'])
    ax3.set_title('Customer Metrics', fontsize=14, fontweight='bold')
    ax3.set_ylabel('Count')
    
    for bar in bars:
        height = bar.get_height()
        ax3.text(bar.get_x() + bar.get_width()/2, height, 
                 f'{int(height)}', ha='center', va='bottom')
    
    # 4. Product Performance
    ax4 = fig.add_subplot(gs[1, 1])
    products = data[['product_a', 'product_b', 'product_c']].sum()
    colors = plt.cm.Set3(np.arange(len(products)))
    wedges, texts, autotexts = ax4.pie(products, labels=products.index, 
                                       autopct='%1.1f%%', colors=colors)
    ax4.set_title('Product Sales Distribution', fontsize=14, fontweight='bold')
    
    # 5. Correlation Heatmap
    ax5 = fig.add_subplot(gs[1, 2])
    numeric_cols = ['revenue', 'customers', 'conversion_rate', 'avg_order_value']
    corr = data[numeric_cols].corr()
    sns.heatmap(corr, annot=True, cmap='RdYlBu', center=0, ax=ax5)
    ax5.set_title('Metrics Correlation', fontsize=14, fontweight='bold')
    
    # 6. Monthly Comparison
    ax6 = fig.add_subplot(gs[2, :])
    width = 0.25
    x = np.arange(len(data['month']))
    
    ax6.bar(x - width, data['product_a'], width, label='Product A')
    ax6.bar(x, data['product_b'], width, label='Product B')
    ax6.bar(x + width, data['product_c'], width, label='Product C')
    
    ax6.set_title('Monthly Product Performance', fontsize=14, fontweight='bold')
    ax6.set_xlabel('Month')
    ax6.set_ylabel('Sales ($)')
    ax6.set_xticks(x)
    ax6.set_xticklabels(data['month'], rotation=45)
    ax6.legend()
    ax6.grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    return fig

# Generate sample report data
report_data = pd.DataFrame({
    'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'revenue': [100000, 120000, 95000, 150000, 140000, 160000],
    'customers': [500, 600, 450, 750, 700, 800],
    'product_a': [30000, 35000, 25000, 45000, 40000, 50000],
    'product_b': [40000, 45000, 35000, 55000, 50000, 55000],
    'product_c': [30000, 40000, 35000, 50000, 50000, 55000],
    'new_customers': [150, 180, 120, 250, 200, 220],
    'returning_customers': [350, 420, 330, 500, 500, 580],
    'churned_customers': [20, 25, 30, 15, 25, 20],
    'conversion_rate': [0.03, 0.035, 0.025, 0.045, 0.04, 0.042],
    'avg_order_value': [200, 200, 210, 200, 200, 200]
})

# Create report
report = create_business_report(report_data)
plt.show()

# Save report
report.savefig('business_report.png', dpi=300, bbox_inches='tight')
print("âœ… Report saved as 'business_report.png'")
```
